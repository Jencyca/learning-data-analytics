***For Excel***

```python
df=pd.read_excel('Customer_data.xlsx','Sheet1')
df
```

```python
def convert_Name_cell(cell):
    if cell =='not available':
        return 'Draco'
    return cell
def convert_Num_of_Items_cell(cell):
    if cell =='n.a.':
        return 8
    return cell
df=pd.read_excel('Customer_data.xlsx','Sheet1',converters={
    'Name':convert_Name_cell,
    'Num_of_Items':convert_Num_of_Items_cell
})
df
```

```python
# to write it in new Excel
df.to_excel('new_customer.xlsx',sheet_name='customer',startrow=1,startcol=2) 
```

```python
# to write two dataframe in one excel ==> meaning 2 sheets in one
df_customer=pd.DataFrame({
    'ID':[1,2,3],
    'Name':['Ron','Hermoine','Harry'],
    'No_of_items':[2,3,4]
})
df_student=pd.DataFrame({
    'ID':[1,2,3],
    'Study_Hours':[5,3,4],
    'Final_Marks':[98,85,92]
})
with pd.ExcelWriter('customer_student.xlsx') as writer:
    df_customer.to_excel(writer,sheet_name='customer')
    df_student.to_excel(writer,sheet_name='student')
```
